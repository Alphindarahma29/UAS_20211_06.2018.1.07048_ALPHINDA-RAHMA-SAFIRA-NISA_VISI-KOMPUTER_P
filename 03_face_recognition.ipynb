```python
import cv2, time
import numpy as np
import os
```

```python
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read('trainer/trainer.yml')
cascadePath = "haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascadePath);

font = cv2.FONT_HERSHEY_SIMPLEX

id = 0

# nama sesuai dengan ids (id sampel)
names = ['Tidak Diketahui', 'Alphinda', 'Rahma', 'Nisa','Andrean']

# menampilkan cam
cam = cv2.VideoCapture(0)
cam.set(3, 1280) # set video widht
cam.set(4, 1024) # set video height

# menentukan min ukuran cam untuk recognation
minW = 0.1*cam.get(3)
minH = 0.1*cam.get(4)

count = 0
count2 = 0

while True:

    ret, img =cam.read()
    #img = cv2.flip(img, -1)

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale( 
        gray,
        scaleFactor = 1.1,
        minNeighbors = 5,
        minSize = (int(minW), int(minH)),
       )

    for(x,y,w,h) in faces:

        cv2.rectangle(img, (x,y), (x+w,y+h), (255,255,255), 2)

        id, confidence = recognizer.predict(gray[y:y+h,x:x+w])

        # jika confidence = 0 maka cocok sempurna 
        if (confidence <= 50):
            count2 = 0
            count += 1

            if (count == 1) :
                print()

            print(f"Wajah Sama {count}")

            id = names[id]
            confidence = " Kecocokan {0}%".format(round(100 - confidence))

            cv2.putText(img, str(id), (x+5,y-5), font, 1, (0,255,0), 3)
            cv2.putText(img, str(confidence), (x+5,y+h-5), font, 1, (0,255,0), 2)

            # recognation berhasil jika telah dicocokkan 30 x
            if (count == 30) :
                print("\nWajah Terdeteksi\n")
                count = 0

        else:
            count = 0
            count2 += 1
            id = names[0]
            cv2.putText(img, str(id), (x+5,y-5), font, 1, (0,0,255), 3)
            
            if (count2 == 30) :
                print("Wajah Tidak Terdeksi")
                count2 = 0

    
    cv2.imshow('camera',img) 

    k = cv2.waitKey(10) & 0xff # Press 'ESC' for exiting video
    if k == 27:
        break

print("\n [INFO] Keluar Dari Program")
cam.release()
cv2.destroyAllWindows()
```
